### Implementing simple Chatbot Using LangGraph

```python
from typing_extensions import TypedDict
from langgraph.graph import START,END,StateGraph

## Reducers
from typing import Annotated
from langgraph.graph.message import add_messages
```

```python
class State(TypedDict):
    messages:Annotated[list,add_messages]
```

```python
from langchain_ollama import ChatOllama

fast_llm = ChatOllama(model='gemma3:270m')
fast_llm.invoke("Hello")
```

```python
slow_llm = ChatOllama(model = "gemma3:1b")
slow_llm.invoke("Hi, I am kalyan")
```

### Now, we will start with Creating Nodes

```python
def super_bot(state:State):
    return {"messages": [slow_llm.invoke(state["messages"])]}
```

```python
graph = StateGraph(state_schema=State)

graph.add_node("SuperBot",super_bot)

graph.add_edge(START,"SuperBot")
graph.add_edge("SuperBot",END)

graph_builder = graph.compile()
```

```python
graph_builder.invoke({"messages": "Hi, I am Kalyan and i like cricket"})
```

```python
graph_builder.invoke({"messages": "Do you know my name?"})
```
